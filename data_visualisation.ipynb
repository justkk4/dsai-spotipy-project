# **Data Visualisation**


```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
songdata = pd.read_csv("datasets/dataset.csv", index_col=0)
songdata
```

## **Numerical Data**

```python
numericaldata = songdata.select_dtypes(include=['int64','float64'])

# drop columns that are of numerical data type but categorical nature
numericaldata.drop(columns=["Time Signature", "Key", "Mode"], axis=1, inplace=True) 
numericaldata
```

```python
numericaldata.describe()
```

```python
f, axes = plt.subplots(5,2,figsize=(20,30))
total = 0
row = 0

audio_and_duration = [c for c in numericaldata.columns if c not in ["Artist Popularity", "Track Popularity"]]

for var in audio_and_duration:
    if total % 2 == 0:
        col = 0
    else:
        col = 1
    
    vp = sb.violinplot(data=numericaldata[var], orient='h', ax=axes[row,col])
    vp.set_xlabel(var, fontsize=10)
    total += 1

    if total % 2 == 0:
        row += 1
```
